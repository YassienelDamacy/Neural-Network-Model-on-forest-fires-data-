## split and normalize the data

```python
y = df['area']
y # make the target is column rain
```

```python
x =df.drop('area',axis=1)
x # the rest is the predictores
```

```python
# Create an instance of StandardScaler
scaler = StandardScaler()
# Fit and transform the dataframe using StandardScaler
x = scaler.fit_transform(x)
# Print the normalized dataframe
print(x)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 50) 
# Split the data into training and testing data
 
#make the train and test data in arrays to enter the ML module
xtrain = np.asanyarray(xtrain)
xtest = np.asanyarray(xtest)
ytrain = np.asanyarray(ytrain)
ytest = np.asanyarray(ytest)
```

```python
num_col = x.shape[1]  # get the columns of the data to be input to Neural networks
num_col
```

## Building Model

```python
def area():
    model = Sequential() # create Sequential model
    model.add(Dense(64, activation='relu', input_shape=(num_col,))) 
    # first hidden layer input is number of columns and activation function is Relu and hidden layer is 64 nodes
    model.add(Dense(32, activation='relu')) # second hidden layer has 32 nodes and activation function is Relu
    model.add(Dense(1)) # output layer is one node

    model.compile(optimizer="adam",loss="mean_squared_error") # use adam optimizer and loss function is MSE
    return model
```

```python
reg = area()
reg.fit(xtrain,ytrain,epochs=50,batch_size=32) # fit model on training data , train on 50 epoch with batch = 32
```

```python
loss = reg.evaluate(xtest,ytest) # test using test data 
```

```python
print("The loss for test set is {}".format(loss))
```

### my opinion on loss is training loss in not high and near test loss so it is not overfitting nor underfitting , if increase hidden layer and number of epochs training loss will decrease and test loss will increase so it will be overfitted , if decrease hidden layer and number of epochs training loss will increase and test loss will increase so it will be underfitted

### so i see it is the best number of hidden layers and number of epochs
